```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import keras
train_df = pd.read_csv("mnist_train.csv", header=None)
test_df = pd.read_csv("mnist_test.csv",header=None)

#output first 10 rows
train_df[:10]
```

```python
#conver pandas dataframe to numpy array
train_matrix = train_df.values
print(train_matrix.shape)
print (train_matrix)
```

```python
train_X = train_matrix[:, 1:]
train_Y = train_matrix[:,:1].ravel(); #ravel converts 2d array into 1d
print (train_X.shape);
print(train_Y.shape);
```

```python
#do the same for test dataset
test_matrix = test_df.values
test_X = test_matrix[:, 1:]
test_Y = test_matrix[:,:1].ravel(); #ravel converts 2d array into 1d
print (test_X.shape);
print(test_Y.shape);
```

```python
def plot_digit(X, y, idx):
    img = X[idx].reshape(28,28)
    plt.imshow(img, cmap='Greys')
    plt.title('%d' % y[idx])
    plt.show()
plot_digit(train_X, train_Y, 23)
```

```python
#Ex 1 multinomial logistic regression
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression(solver='lbfgs', multi_class = 'multinomial')

log_model.fit(train_X, train_Y)

pred_Y = log_model.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score
print ('Test accuracy:', accuracy_score(pred_Y, test_Y))
```
